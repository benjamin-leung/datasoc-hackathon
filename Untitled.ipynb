```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/train.csv')

# sa = df["sex and age"].str.split(",", n = 1, expand = True) 
# jsla = df["job_status and living_area"].str.split("?", n = 1, expand = True) 

jsla = df[["JOB_STATUS", "LIVING_AREA"]].copy()

# df.drop(columns = ["sex and age"], inplace = True)
# df.drop(columns = ["job_status and living_area"], inplace = True)

def swap(ls, j):
    first = str(ls["JOB_STATUS"][j])
    second = str(ls["LIVING_AREA"][j])
    ls["JOB_STATUS"][j] = second
    ls["LIVING_AREA"][j] = first
    
# print(jsla)
    
for i in range(len(jsla)):
#     jsla["JOB_STATUS"][i] = str(jsla["JOB_STATUS"][i]).upper()
#     jsla["LIVING_AREA"][i] = str(jsla[1][i]).upper()
    if(isinstance(jsla["JOB_STATUS"][i], str)):
        if(len(jsla["JOB_STATUS"][i]) != 0):
            if(jsla["JOB_STATUS"][i][0] == "R" or jsla["JOB_STATUS"][i][0] == "C"):
#                 print(i)
                swap(jsla, i)



df.drop(columns = ["JOB_STATUS"], inplace = True)
df.drop(columns = ["LIVING_AREA"], inplace = True)

df["job_status"] = jsla["JOB_STATUS"]
df["living_status"] = jsla["LIVING_AREA"]

# df["job_status"]
# df.head(10)

# print(jsla["JOB_STATUS"][29])

# jsla

df

# print(df["high_BP"].head(50))
```
