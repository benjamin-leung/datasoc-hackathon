```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv('unbalanced_mice.csv')
train.head()
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
```

```python
job_status = pd.get_dummies(train['job_status'],drop_first=True)
```

```python
living_area = pd.get_dummies(train['living_area'],drop_first=True)
```

```python
smoker_status = pd.get_dummies(train['smoker_status'],drop_first=True)
```

```python
train.drop(['Sex','job_status','living_area','smoker_status', 'id', 'TreatmentA', 'TreatmentB', 'TreatmentC', 'TreatmentD', 'average_blood_sugar'],axis=1,inplace=True)
```

```python
train = pd.concat([train, smoker_status],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train.drop('stroke_in_2018',axis=1), 
                                                    train['stroke_in_2018'], test_size=0.3, 
                                                    random_state=101)
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(y_train.astype(int),X_train.astype(float))
result=logit_model.fit()
print(result.summary2())
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions)) 
```
