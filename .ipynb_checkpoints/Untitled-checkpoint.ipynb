```python
import pandas as pd

df = pd.read_csv('data/train.csv')

sa = df["sex and age"].str.split(",", n = 1, expand = True) 
jsla = df["job_status and living_area"].str.split("?", n = 1, expand = True) 

df.drop(columns = ["sex and age"], inplace = True)
df.drop(columns = ["job_status and living_area"], inplace = True)

def swap(ls, j):
    first = str(ls[0][j])
    second = str(ls[1][j])
    ls[0][j] = second
    ls[1][j] = first

# def upperCol(ls, j):
#     print(ls[j])
#     for i in range(len(ls)):
#         if(isinstance(ls[j][i], str)):
#             if(len(ls[j][i]) != 0):
#                 print(ls[j][i])
#                 ls[j][i] = str(ls[j][i]).upper()
    
    
# print(jsla[0].unique())
    
for i in range(len(jsla)):
    if(isinstance(jsla[0][i], str)):
        if(len(jsla[0][i]) != 0):
            if(jsla[0][i][0].upper() == "R" or jsla[0][i][0].upper == "C"):
                swap(jsla, i)

# print(jsla)

df["sex"] = sa[0]
df["age"] = sa[1]
df["job_status"] = jsla[0]
df["living_status"] = jsla[1]

# upperCol(df, "job_status")
# df["job_status"]
df.head()
```
